<h3>Training and Visualizing a decision Tree</h3>

```python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris['data'][:,2:], iris['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_clf = DecisionTreeClassifier(max_depth = 2)
dec_clf.fit(X,y)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dec_clf,out_file='./iris_tree.dot', 
                feature_names=iris['feature_names'][:2],
                class_names=iris['target_names'],
                rounded=True,
                filled=True)
```

```python
#After Exporting image using this command $ dot -Tpng iris_tree.dot -o iris_tree.png
```

<h3> Estimating the Probabilities </h3>

```python
dec_clf.predict_proba([[5,1.5]])
```

<h3>Compuational Complexity</h3>

<ol>
   <li>Making Decision:->O(log(m)) i.e height of tree</li>
   <li>Building a Tree:->O(n*m*log(m))</li>

</ol>

<h3>Gini Impurity or Entropy</h3>
    we can either of those, default is gini, you set to entropy using <b>criterion</b> hyperparam

```python
dec_entropy_clasf = DecisionTreeClassifier(criterion='entropy', max_depth=2)
dec_entropy_clasf.fit(X,y)
dec_entropy_clasf.predict_proba([[5,1.5]])
```

```python
#chaning max depth gives different answer
dec_entropy_clasf = DecisionTreeClassifier(criterion='entropy')
dec_entropy_clasf.fit(X,y)
dec_entropy_clasf.predict_proba([[5,1.5]])
```
